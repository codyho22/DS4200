```python
import altair as alt
import pandas as pd
import numpy as np
```

```python
# final list of Enslavers who owned 2 or more slaves

data = {'Enslaver':['John Alford', 'Austin Ebenezer', 'John Dowse', 'Joseph Lynde', 'William Wyer', 'Robert Temple', 'Jonathan Dowse', 'Hammond Gowin', 'Captain Rows', 'Thomas Russell'],
 'Occupation':['Captain', 'Silversmith', 'Captain', 'Merchant', 'Sail Maker', 'Captain', 'Captain', 'Captain', 'Captain', 'Merchant'],
 'Number_of_Enslaved':[3, 3, 3, 3, 3, 6, 2, 2, 2, 2]}

#convert int pd df
df = pd.DataFrame(data)

df
```

```python
# Get counts of each occupation
occupation_counts = df.groupby('Occupation').size().reset_index(name='count')
occupation_counts
```

```python
alt.Chart(occupation_counts).mark_bar().encode(
    x=alt.X('Occupation:N', axis=alt.Axis(labelAngle=-45, labelPadding=10)),    
    y=alt.Y('count:Q', title='Count of Individuals', axis=alt.Axis(tickCount=6, labelPadding=10))
).properties(
    width=500,
    height=400,
    title='1775 Charlestown: Male Occupations owning 2 or More Slaves'
).configure_axis(
    labelFontSize=14,
    titleFontSize=16
).configure_title(
    fontSize=20
)
    
```
