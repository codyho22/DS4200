```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv('Summer Olympic Medals.csv')

df
```

```python
df_sorted = df.sort_values(by='Total', ascending=False)


df_long = df_sorted.melt(id_vars=['NOC'], value_vars=['Gold', 'Silver', 'Bronze'],
                  var_name='medal', value_name='count')

# Create the stacked bar chart
chart = alt.Chart(df_long).mark_bar().encode(
    x=alt.X('NOC:N', title='Country'),
    y=alt.Y('count:Q', title='Number of Medals'),
    color=alt.Color('medal:N', title='Medal Type', 
                    scale=alt.Scale(domain=['Gold', 'Silver', 'Bronze'],
                                    range=['#FFD700', '#C0C0C0', '#CD7F32']))
).properties(
    title='Olympic Medals by Country',
    width=1000,
    height=400
)

# Display the chart
chart
```
