```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv('Summer Olympic Medals.csv')

df
```

```python
lst_golds = df['Gold']

length = len(lst_golds)
gold_list = ['Gold'] * length
silver_list = ['Silver'] * length
bronze_list = ['Bronze'] * length





gold_df = df[['NOC', 'Gold']]
gold_df['Medal'] = gold_list
gold_df.rename(columns={'NOC': 'Country', 'Gold': 'Count'}, inplace=True)


silver_df = df[['NOC', 'Silver']]
silver_df['Medal'] = silver_list
silver_df.rename(columns={'NOC': 'Country', 'Silver': 'Count'}, inplace=True)

bronze_df = df[['NOC', 'Bronze']]
bronze_df['Medal'] = bronze_list
bronze_df.rename(columns={'NOC': 'Country', 'Bronze': 'Count'}, inplace=True)
```

```python
combined_df = pd.concat([gold_df, silver_df], ignore_index=True)

combined_df = pd.concat([combined_df, bronze_df], ignore_index=True)

combined_df
```
