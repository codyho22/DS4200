```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv('Summer Olympic Medals.csv')

df
```

```python
barchart = alt.Chart(df).mark_bar().encode(
    x='Total:Q',
    y='NOC'
)

barchart.encode(color='Gold')
```

```python
brush_legend = alt.selection_point(
    fields = ["medal"],
    bind="legend" 
)



df_long = df.melt(id_vars=['NOC'], value_vars=['Gold', 'Silver', 'Bronze'],
                  var_name='medal', value_name='count')

chart = alt.Chart(df_long).mark_bar(tooltip=True).encode(
    y=alt.X('NOC:N', title='Country'),
    x=alt.Y('count:Q', title='Number of Medals'),
    color=alt.condition(brush_legend, 'medal:N', alt.value('lightgray'), 
                        title='Medal Type',
                        scale=alt.Scale(domain=['Gold', 'Silver', 'Bronze'],
                                        range=['#FFD700', '#C0C0C0', '#CD7F32']))
).properties(
    title='Olympic Medals by Country',
    width=600,
    height=1500
).add_params(brush_legend)

chart
```
