```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv('Summer Olympic Medals.csv')

df
```

```python
lst_golds = df['Gold']

length = len(lst_golds)
gold_list = ['Gold'] * length
silver_list = ['Silver'] * length
bronze_list = ['Bronze'] * length





gold_df = df[['NOC', 'Gold']]
gold_df['Medal'] = gold_list
gold_df.rename(columns={'NOC': 'Country', 'Gold': 'Count'}, inplace=True)


silver_df = df[['NOC', 'Silver']]
silver_df['Medal'] = silver_list
silver_df.rename(columns={'NOC': 'Country', 'Silver': 'Count'}, inplace=True)

bronze_df = df[['NOC', 'Bronze']]
bronze_df['Medal'] = bronze_list
bronze_df.rename(columns={'NOC': 'Country', 'Bronze': 'Count'}, inplace=True)
```

```python
combined_df = pd.concat([gold_df, silver_df], ignore_index=True)

combined_df = pd.concat([combined_df, bronze_df], ignore_index=True)

combined_df
```

```python
brush_legend = alt.selection_point(
    fields = ["Medal"],
    bind="legend" 
)

brush_bar = alt.selection_interval()


stacked_chart = alt.Chart(combined_df).mark_bar(tooltip=True).encode(
    x=alt.X('sum(Count):Q', title='Count of Total Medals'),
    y=alt.Y('Country:O', title='Country Name'),
    color=alt.condition(brush_legend, 
            'Medal',
             alt.value('lightgray'), scale=alt.Scale(domain=['Gold', 'Silver', 'Bronze'],
                                        range=['#FFD700', '#C0C0C0', '#CD7F32']))
).add_params(
    brush_legend, brush_bar
).properties(
    title='Barchart Displaying Summer Olympic Medals by Type')

#interactivity: tooltip, brush legend, brush bar, slider
```

```python
slider = alt.binding_range(min=1, max=3000, step=1, name='Cutoff')
selector = alt.selection_point(
    name="SelectorName",
    fields=['cutoff'],
    bind=slider,
    value=3000
)

slider_chart = alt.Chart(df).mark_point().encode(
    x=alt.X('Total:Q', title='Count of Total Medals'),
    y=alt.Y('NOC:N', title='Country Name'),
    color=alt.condition(
        alt.datum.Total <= selector.cutoff,
        alt.value('red'),
        alt.value('white')
    )
).properties(
    title='Total Summer Olympic Medals by Country'
).add_params(
    selector
)
```
