```python
import altair as alt
import pandas as pd
import numpy as np
```

```python
data = {'Name':['John Alford', 'Austin Ebenezer', 'John Dowse', 'Joseph Lynde', 'William Wyer', 'Robert Temple'],
 'Occupation':['Captain', 'Silversmith', 'Captain', 'Merchant', 'Sail Maker', 'Captain'],
 'Number_of_Enslaved':[3, 3, 3, 3, 3, 6]}

df = pd.DataFrame(data)

df
```

```python
slave_occupation_df = pd.DataFrame({'Count':[3, 1, 1, 1],
 'Occupation':['Captain', 'Silversmith', 'Merchant', 'Sail Maker'],
 'Avg_Number_of_Enslaved':[4.5, 3, 3, 3],
 'Avg_Age_Enslaver': [51, 42, 73, 67],
 'Num_Slaves': [12, 3, 3, 3]})

slave_occupation_df
```

```python
brush_legend = alt.selection_point(
    fields = ["Occupation"],
    bind="legend" 
)

bubble_chart = alt.Chart(slave_occupation_df).mark_circle(tooltip=True).encode(
    x='Occupation:N',    
    y=alt.Y('Avg_Age_Enslaver:Q', title='Average Age of Enslaver'),
    size='Count:Q',
    color=alt.condition(brush_legend, 'Occupation:N', alt.value('lightgray')),
    tooltip=['Count', 'Avg_Number_of_Enslaved']
).configure_axis(
    grid=False
).properties(
    width=300,
    height=200,
    title='Bubble Chart of Male Occupations owning 3 or More Slaves'
).add_params(
    brush_legend
)


bubble_chart
```
