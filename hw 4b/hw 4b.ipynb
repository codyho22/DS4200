```python
import pandas as pd
import altair as alt
```

```python
# input the values from the visualization I am redesigning
x = list(range(2022, 2033))

y = [2.09, 2.44, 2.85, 3.33, 3.89, 4.55, 5.31, 6.2, 7.25, 8.47, 9.89]

#use the x and y lists and combine into pd df
df = pd.DataFrame({'x': x, 
                   'y':y})

df
```

```python
#create barchart in altair
barChart = alt.Chart(df, title=alt.Title("Model's Prediction for AI Training Dataset Market Size, 2023 to 2032")).mark_bar(size=30).encode(
    x=alt.X("x:O", title='Year'),
    y=alt.Y("y:Q", title='Dataset Marketsize (USD Billion)') 
).properties(
    width=600,
    height=500)

#create text labels
text = barChart.mark_text(align= 'center', dy=-10).encode(text='y:Q')

#overlay the text labels with the barchart into chart
chart = (barChart + text)

#display
chart
```
